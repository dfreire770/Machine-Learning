```python
import pandas as pd

train = pd.read_csv('train.csv')
```

```python
train = train[['wav_filename','transcript']]
```

```python
def build_data(audio_directory, data):
    
    dataset = {}
    for index, row in data.iterrows():
        file_name = row['wav_filename']
        transcript = row['transcript']
        feature = extract_mfcc(str(audio_directory + file_name))
        if file_name not in dataset.keys():
            dataset[file_name] = []
            dataset[file_name].append(feature) #file_name = label
            dataset[file_name].append(transcript)
        else:
            exist_feature = dataset[file_name]
            exist_feature.append(feature)
            exist_feature.append(transcript)
            dataset[file_name] = exist_feature
        
    return dataset
    

filesList = train[:100]
audio_directory = './clips/'

build_data(audio_directory, filesList)

```

```python
def train_GMMHMM(dataset):
    GMMHMM_Models = {}
    states_num = 5
    GMM_mix_num = 6
    tmp_p = 1.0/(states_num-2)
    transmatPrior = np.array([[tmp_p, tmp_p, tmp_p, 0 ,0], \
    [0, tmp_p, tmp_p, tmp_p , 0], \
    [0, 0, tmp_p, tmp_p,tmp_p], \
    [0, 0, 0, 0.5, 0.5], \
    [0, 0, 0, 0, 1]],dtype=np.float)
    startprobPrior = np.array([0.5, 0.5, 0, 0, 0],dtype=np.float)
    for label in dataset.keys():
        model = hmm.GMMHMM(n_components=states_num, n_mix=GMM_mix_num, \
        transmat_prior=transmatPrior, startprob_prior=startprobPrior, \
        covariance_type='diag', n_iter=10)
        trainData = dataset[label]
        length = np.zeros([len(trainData), ], dtype=np.int)
    for m in range(len(trainData)):
        length[m] = trainData[m].shape[0]
        trainData = np.vstack(trainData)
        model.fit(trainData, lengths=length)
        GMMHMM_Models[label] = model
    return GMMHMM_Models
```

```python
def main():
    trainDir = './clips/'
    train_filesList = train[:100]
    trainDataSet = build_data(trainDir, train_filesList)
    
    print("Finish prepare the training data")
    
    hmmModels = train_GMMHMM(trainDataSet)
    
    print("Finish training of the GMM_HMM models")
    
    testDir = './clips/'
    testDataSet = build_data(testDir)
    score_cnt = 0
    for label in testDataSet.keys():
        feature = testDataSet[label]
        scoreList = {}
        for model_label in hmmModels.keys():
            model = hmmModels[model_label]
            score = model.score(feature[0])
            scoreList[model_label] = score
            predict = max(scoreList, key=scoreList.get)
            print("Test on true label ", label, ": predict result label is ", predict)
            if predict == label:
                score_cnt+=1
    print("Final recognition rate is %.2f"%(100.0*score_cnt/len(testDataSet.keys())), "%")
```
